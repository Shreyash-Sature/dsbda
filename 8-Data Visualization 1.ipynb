```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("titanic.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
```

```python
df['Cabin']=df['Cabin'].fillna(df['Cabin'].ffill())
```

```python
df['Embarked']=df['Embarked'].fillna(df['Embarked'].ffill())
```

```python
df.isnull().sum()
```

```python
df['Cabin']=df['Cabin'].fillna(df['Cabin'].bfill())
```

```python
df.isnull().sum()
```

```python
sns.histplot(df['Fare'],bins=10)
```

```python
sns.histplot(df["Fare"], bins=10, kde=True)
plt.title("Fare distribution with KDE")
plt.show()
```

```python
sns.boxplot(x='Sex', y='Age', data=df)
plt.show()
```

```python
sns.countplot(x="Sex", data=df)
plt.title("Gender based distribution")
plt.show()
```

```python
sns.countplot(x="Sex", hue="Survived", data=df)
plt.title("Gender based survival distribution")
plt.show()
```
