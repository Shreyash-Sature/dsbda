```python
import nltk
import re
import pandas as pd

from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
```

```python
# 1. Tokenization
# -------------------------------
text1 = "Tokenization is the first step in text analytics."

print("Sentence Tokenization:")
print(sent_tokenize(text1))

print("\nWord Tokenization:")
print(word_tokenize(text1))
```

```python
# 2. Stopword Removal
# -------------------------------
text2 = "How to remove stop words with NLTK library in Python?"

stop_words = set(stopwords.words("english"))

# Remove special characters
text2 = re.sub('[^a-zA-Z]', ' ', text2)

tokens = word_tokenize(text2.lower())

filtered_words = [w for w in tokens if w not in stop_words]

print("\nAfter Stopword Removal:")
print(filtered_words)
```

```python
# 3. Stemming
# -------------------------------
ps = PorterStemmer()
words = ["studies", "studying", "cries", "cry"]

print("\nStemming:")
for w in words:
    print(ps.stem(w))
```

```python
#4. Lemmatization
# -------------------------------
lemmatizer = WordNetLemmatizer()

print("\nLemmatization:")
for w in words:
    print(lemmatizer.lemmatize(w))
```

```python
# 5. POS Tagging
# -------------------------------
sentence = "The pink sweater fit her perfectly"

tokens = word_tokenize(sentence)

print("\nPOS Tagging:")
print(pos_tag(tokens))
```

```python
# 6. TF-IDF Vectorization
# -------------------------------
docs = [
    "Jupiter is the largest planet",
    "Mars is the fourth planet from the sun"
]

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(docs)

print("\nFeature Names:")
print(vectorizer.get_feature_names_out())

print("\nTF-IDF Matrix:")
print(X.toarray())
```
