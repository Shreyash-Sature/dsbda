```python
min_val = df['SepalLengthCm'].min()
max_val = df['SepalLengthCm'].max()
df['SepalLength_Norm'] = (df['SepalLengthCm'] - min_val) / (max_val - min_val)
print(df['SepalLength_Norm'])
```
