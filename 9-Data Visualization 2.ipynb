```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("titanic.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
df['Cabin']=df['Cabin'].fillna(df['Cabin'].ffill())
df['Embarked']=df['Embarked'].fillna(df['Embarked'].ffill())
```

```python
df.isnull().sum()
```

```python
df['Cabin']=df['Cabin'].fillna(df['Cabin'].bfill())
```

```python
df.isnull().sum()
```

```python
sns.boxplot(data=df,x='Sex',y='Age',hue='Survived')
```
